## CARGA DE IMAGENES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def mostrar_imagenes(img_bgr, titulo):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    img_false_color = cv2.merge([img_gray, img_gray, img_gray])
    
    alto, ancho, _ = img_bgr.shape
    
    b, g, r = cv2.split(img_bgr)
    neg_b = 255 - b
    neg_g = 255 - g
    neg_r = 255 - r
    
    fig, axs = plt.subplots(2, 3, figsize=(15, 10))
    
    axs[0, 0].imshow(img_rgb)
    axs[0, 0].set_title("RGB")
    axs[0, 0].axis("on")
    
    axs[0, 1].imshow(img_gray, cmap="gray")
    axs[0, 1].set_title("Escala de grises")
    axs[0, 1].axis("on")
    
    ##axs[0, 2].imshow(img_false_color)
    ##axs[0, 2].set_title("Color falso")
    ##axs[0, 2].axis("off")
    
    axs[0, 2].imshow(neg_r, cmap="Reds")
    axs[0, 2].set_title("falso Rojo")
    axs[0, 2].axis("on")
    
    axs[1, 0].imshow(neg_g, cmap="Greens")
    axs[1, 0].set_title("falso Verde")
    axs[1, 0].axis("on")
    
    axs[1, 1].imshow(neg_b, cmap="Blues")
    axs[1, 1].set_title("falso Azul")
    axs[1, 1].axis("on")
    
    plt.suptitle(f"{titulo} ")
    plt.show()

    cv2.imwrite(f"{titulo}_BGR.png", img_bgr)
    cv2.imwrite(f"{titulo}_Gris.png", img_gray)
    cv2.imwrite("imagen2.png", img_bgr)
    plt.imsave("imagen3.png", img_rgb)



img1 = cv2.imread("images1.jpg")
mostrar_imagenes(img1, "Imagen 1")
```

## IMAGEN 2

```python
img2 = cv2.imread("images2.jpg")
mostrar_imagenes(img2, "Imagen 2")
```
